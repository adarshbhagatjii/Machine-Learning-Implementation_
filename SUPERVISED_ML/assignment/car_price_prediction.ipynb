Car price prediction 


```python
import pandas as pd
```

```python
data=pd.read_csv('car_price_prediction_with_missing.csv')
data
```

```python
data.isna().sum()
```

```python
data=data.dropna()
data.isna().sum()
```

```python
data
```

```python
df=pd.get_dummies(data,drop_first=True, dtype=int)
df
```

```python
# x=df.drop('Price',axis=1)
# y=df.Price
# x
x=df[['Year', 'Engine Size', 'Mileage']]
y=df['Price']
```

```python
# from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix,f1_score,recall_score,precision_score
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
# knrmodel=KNeighborsRegressor(n_neighbors=5)
# knrmodel.fit(xtrain,ytrain)
model=LinearRegression()
# model=Lasso()
model.fit(xtrain,ytrain)
```

```python
ypred=model.predict(xtest)
ypred
```

```python
print('training score:', model.score(xtrain,ytrain))
print('testing score:', model.score(xtest,ytest))
```
