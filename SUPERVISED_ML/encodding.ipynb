```python
import pandas as pd
```

```python
# label encodding 
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df=pd.DataFrame({'Gender':['M','F','M','M','F']})
df
```

```python
df['Gender']=le.fit_transform(df['Gender'])
df
```
