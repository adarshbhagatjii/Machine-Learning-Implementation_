```python
import pandas as pd
from sklearn.datasets import load_breast_cancer, fetch_california_housing
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix,f1_score,recall_score,precision_score, r2_score, mean_absolute_error
```

```python
cdata=load_breast_cancer()
```

```python
df=pd.DataFrame(cdata.data,columns=cdata.feature_names)
df['cancer']=cdata.target 
df
```

```python
x=df.drop('cancer',axis=1)
y=df['cancer']
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
knn_model=KNeighborsClassifier(n_neighbors=10)
knn_model.fit(xtrain,ytrain)
```

```python
ypred=knn_model.predict(xtest)
ypred
```

```python
print('training score:', knn_model.score(xtrain,ytrain))
print('testing score:', knn_model.score(xtest,ytest))
```

```python
print("Accuracy:", accuracy_score(ytest, ypred))
print("Precision:", precision_score(ytest, ypred))
print("Recall:", recall_score(ytest, ypred))
print("F1 Score:", f1_score(ytest, ypred))
```

```python
print('mean_absolute_error:', mean_absolute_error(ytest, ypred))
print('confusion_matrix:\n', confusion_matrix(ytest, ypred))
```

```python
rdata=fetch_california_housing()
rdf=pd.DataFrame(rdata.data,columns=rdata.feature_names)
rdf['MedHouseVal']=rdata.target
rdf
```

```python
x=rdf.drop('MedHouseVal',axis=1)
y=rdf['MedHouseVal']
```

```python
knn_reg=KNeighborsRegressor(n_neighbors=5)
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=42)
knn_reg.fit(xtrain,ytrain)
```

```python
ypred=knn_reg.predict(xtest)
ypred
```

```python
print('training score:', knn_reg.score(xtrain,ytrain))
print('testing score:', knn_reg.score(xtest,ytest))
```

```python
print('r2_score:', r2_score(ytest, ypred))
print('mean_absolute_error:', mean_absolute_error(ytest, ypred))
```
