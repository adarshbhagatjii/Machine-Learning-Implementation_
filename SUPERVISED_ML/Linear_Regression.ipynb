```python
import pandas as pd 
from sklearn import datasets
from sklearn.datasets import fetch_california_housing
```

```python
data=fetch_california_housing()
data
```

#creating dataframe from the dataset provided

```python
#  create a a dataframe 

df=pd.DataFrame(data.data, columns=data.feature_names)
df
```

```python
#creating output column
# df['price']= data.target
# df
df[data.target_names[0]]=data.target
df
```

```python
#dividing inputs and outputs
# X= df.iloc[:, :-1] #-1 because, total columns=9, from column 1 till 9-1 it will take as input or X
# X
X= df.drop('MedHouseVal', axis=1) #dropping the output column gives us onlly the input ones
X
```

```python
y= df['MedHouseVal'] #giving the output column only
y
```

```python
#dividing the dataset into training and testing, so that if we train the dataset with all the data available we wont have any data to test the model. so we divide it into 2, as per 70 and 30
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size=0.3, random_state=42) #random_state=42, so that we have only one set of testing values so that if we refresh, the values wont change
```

```python
from sklearn.linear_model import LinearRegression
model= LinearRegression()
model
```

```python
dir(LinearRegression)
```

```python
#fit is used to train the model
model.fit(X_train, y_train)
```

```python
#after the model is trained, i need to find the predictions on the testing dataset
y_pred= model.predict(X_test)
y_pred
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
#after the predictions are done, checking the accuracy using performance matrices
r2= r2_score(y_test, y_pred)
r2
```

```python
#checking adjusted r2score
N= X_test.shape[0]
P= X_test.shape[1]
adjr2= 1-((1-r2)*(N-1)/(N-P-1))
adjr2
```

```python
#checking mean squared error
mean_squared_error(y_test, y_pred)
```

```python
#checking mean absolute error
mean_absolute_error(y_test, y_pred)
```

```python
#checking if the model is overfitting or underfitting
print('Training: ', model.score(X_train, y_train))
print('Testing: ', model.score(X_test, y_test))
```
