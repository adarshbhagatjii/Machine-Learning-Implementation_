```python
import pandas as pd 
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier, GradientBoostingRegressor
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
```

```python
data =pd.read_csv('F:\QSP\ML_QSP\ML_WORKFLOW\data\heart.csv')
data.head()
```

```python
data.info()
data.drop(data.columns[0],axis=1,inplace=True)
data.head()
```

```python
x= data.drop('target',axis=1)
# x= data.drop(data.columns[0,-1],axis=1)
y= data['target']
x
```

```python
xtrain, xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3, random_state=42)
```

```python
gbmodel=GradientBoostingClassifier(n_estimators=150, learning_rate=0.005)
```

```python
gbmodel.fit(xtrain,ytrain)
```

```python
ypred=gbmodel.predict(xtest)
ypred
```

```python
print('training efficiency', gbmodel.score(xtrain,ytrain))
print('testing efficiency', gbmodel.score(xtest,ytest))
```

```python
from sklearn.datasets import fetch_california_housing, load_diabetes
from sklearn.preprocessing import StandardScaler
```

```python
data=fetch_california_housing()
df=pd.DataFrame(data.data, columns=data.feature_names)
df[data.target_names[0]]=data.target
df
```

```python
x= df.drop('MedHouseVal', axis=1)
y= df['MedHouseVal']
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.3, random_state=42)
scaler=StandardScaler()
xtrain=scaler.fit_transform(xtrain)
xtest=scaler.transform(xtest)
```

```python
Gbrmodel=GradientBoostingRegressor(n_estimators=150, learning_rate=0.1)
Gbrmodel.fit(xtrain,ytrain)
```

```python
ypred=Gbrmodel.predict(xtest)
ypred
```

```python
print('training efficiency', Gbrmodel.score(xtrain,ytrain))
print('testing efficiency', Gbrmodel.score(xtest,ytest))
```
