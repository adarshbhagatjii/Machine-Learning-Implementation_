```python
import pandas as pd 
from sklearn.svm import SVC, SVR
from sklearn.model_selection import train_test_split
from sklearn.metrics import recall_score, precision_score, f1_score, accuracy_score, r2_score, mean_absolute_error, mean_squared_error 
```

```python
data =pd.read_csv('data/bankloan.csv')
data.head()
```

```python
data.drop(['ZIP.Code','ID'], axis=1, inplace=True)
data
```

```python
x=data.drop('Personal.Loan', axis=1)
y=data['Personal.Loan']
```

```python
xtrain, xtest, ytrain, ytest =train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
clsSvcModel =SVC(kernel='rbf')
clsSvcModel.fit(xtrain, ytrain)
```

```python
ypred=clsSvcModel.predict(xtest)
ypred
```

```python
print('training score :', clsSvcModel.score(xtrain, ytrain))
print('testing score :', clsSvcModel.score(xtest, ytest))
```
