## Car category classification using Random Forest Classifier

```python
import pandas as pd 
import numpy as np
```

```python
data=pd.read_csv('data/global_cars_enhanced.csv')
data.head()
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, OrdinalEncoder
```

```python
data.info()
```

```python
data=data.drop(['Car_ID'], axis=1)
```

```python
data['Price_Category'].value_counts()
```

```python
oe= OrdinalEncoder(categories=[['Budget', 'Mid-Range', 'Premium','Luxury']])
data['Price_Category']=oe.fit_transform(data[['Price_Category']])
data.head()
```

```python
le= LabelEncoder()
data['Age_Category']=le.fit_transform(data['Age_Category'])
data.head()
```

```python
dff= pd.get_dummies(data, dtype=int)
dff.head()
```

```python
x=dff.drop('Price_Category', axis=1)
y=dff['Price_Category']
x.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain ,xtest,ytrain,ytest= train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score, precision_score, recall_score
```

```python
RFCModel=RandomForestClassifier(n_estimators=100, max_depth=5)
RFCModel.fit(xtrain,ytrain)
```

```python
ypred=RFCModel.predict(xtest)
```

```python
print('training efficiency:', RFCModel.score(xtrain,ytrain))
print('testing efficiency:', RFCModel.score(xtest, ytest))
```

```python
print('confusion_matrix\n', confusion_matrix(ytest, ypred))
```

```python
print('accuracy_score',accuracy_score(ytest,ypred))
print('precision_score',precision_score(ytest,ypred, average='weighted'))
print('recall_score', recall_score(ytest,ypred, average='weighted'))
print('f1_score', f1_score(ytest,ypred, average='weighted'))
```
