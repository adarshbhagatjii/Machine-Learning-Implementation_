```python
import matplotlib.pyplot as plt 
from sklearn.cluster import DBSCAN, KMeans, AgglomerativeClustering
```

```python
from sklearn.datasets import make_moons, load_iris
```

```python
x,y=make_moons(n_samples=1000, noise=0.05, random_state=42)
```

```python
plt.scatter(x[:,0],x[:,1], c='black')
```

```python
model=DBSCAN(eps=0.08,min_samples=5)
model.fit(x)
```

```python
label=model.labels_
```

```python
plt.scatter(x[:,0],x[:,1], c=label)
```

```python
kmodel=KMeans(n_clusters=2)
kmodel.fit(x)
```

```python
leb=kmodel.labels_ 
```

```python
plt.scatter(x[:,0],x[:,1], c=leb)
```

```python
hmodel=AgglomerativeClustering(n_clusters=2)
hmodel.fit(x)
```

```python
lebl=hmodel.labels_
```

```python
plt.scatter(x[:,0],x[:,1], c=lebl,cmap='rainbow')
```

```python
data=load_iris()
x=data.data[:,2:4]
```

```python
plt.scatter(x[:,0],x[:,1], c='black')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
x_scaled=scaler.fit_transform(x)
```

```python
dbmodel=DBSCAN(eps=0.5, min_samples=8)
dbmodel.fit(x_scaled)
```

```python
labels=dbmodel.labels_
```

```python
plt.scatter(x_scaled[:,0],x_scaled[:,1], c=labels, cmap='rainbow')
```
