```python
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.metrics import silhouette_score
```

```python
wine = load_wine()
X = wine.data
```

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X)
```

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1])
```

```python
plt.scatter(X_scaled[:, wine.feature_names.index('alcohol')], X_scaled[:, wine.feature_names.index('malic_acid')])
```

```python

linked=linkage(X_scaled, method='ward')
plt.figure(figsize=(12,6))
dendrogram(linked)
plt.title('Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
```

```python
model=AgglomerativeClustering(n_clusters=3,linkage='ward')
labels=model.fit_predict(X_scaled)
labels
```

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1],c=labels, cmap='rainbow')
plt.title('Wine Dataset (Scaled)')
plt.xlabel('alcohal')
plt.ylabel('malic_acid')
```

```python
print("Silhouette Score:", silhouette_score(X_scaled, labels))
```

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
X_pca=pca.fit_transform(X_scaled)
```

```python
model=AgglomerativeClustering(n_clusters=3)
```

```python
labels_pca=model.fit_predict(X_pca)
labels_pca
```

```python
silhouette_score(X_pca, labels_pca)
```
